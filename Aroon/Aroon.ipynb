```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

```python
path = 'Z:\data\BTCUSDT\BTCUSDT-1h'
```

```python
filelist = os.listdir(path)
filelist.sort()
df = pd.DataFrame()
for file in tqdm(filelist[-400:]):
    df1 = pd.read_csv(os.path.join(path, file))
    df = pd.concat([df, df1])
df
```

```python
df1 = df[['open_time', 'open', 'high', 'low', 'close', 'volume']]
df1.columns = ['t', 'o', 'h', 'l', 'c', 'v']
df1 = df1.dropna()
df1
```

```python
N = 25  # 默认周期为25天
df1['g1.Aroon_Up'] = ((N - df1['c'].rolling(N).apply(lambda x: N - x.argmax()))/N) *100
df1['g1.Aroon_Down'] = ((N - df1['c'].rolling(N).apply(lambda x: N - x.argmin()))/N) *100
df1=df1[30:]
```

```python
df1.to_csv('./AroonH.csv', index=False)
```
